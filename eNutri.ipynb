```python
import re

def _read_points(df_hasil):
    all_unique_activities = df_hasil['Aktivitas_Bidan'].unique()
    activities_with_points = [unique for unique in all_unique_activities if 'point' in unique]

    columns_to_keep = ['Tanggal', 'Bulan', 'Provinsi', 'Nama_Bidan', 'Keterangan']

    data_points = df_hasil.loc[df_hasil['Aktivitas_Bidan'].isin(activities_with_points), columns_to_keep]

    data_points['Perolehan_Point'] = data_points['Keterangan'].apply(lambda row: re.findall(r'\+(\d+)\s', row)[0]).astype('int')

    return data_points

new_df = _read_points(df_hasil)
new_df
```

```python
from pandas.io import gbq

new_df.to_gbq(destination_table='eNutri.point-set',
            project_id= 'enutri-359007',
            if_exists= 'replace')
```
